```python
import numpy as np
```

```python
#ID array
arr1 = np.array([1, 2, 3, 4, 5])

print("1D Array:")
print(arr1)
```

```python
#2D Array
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("2D Array:")
print(arr2)
```

```python
#3D Array
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("2D Array:")
print(arr2)
```

```python
print("Shape of arr1:", arr1.shape)
print("Shape of arr2:", arr2.shape)
```

```python
arr3 = np.array([1, 2, 3, 4, 5, 6])

reshaped_arr = arr3.reshape(2, 3)

print("Reshaped Array:")
print(reshaped_arr)
```

```python
print("First Element:", arr1[0])
print("Elements from index 1 to 3:", arr1[1:4])
```

```python
print("Sum:", np.sum(arr1))
```

```python
print("Mean:", np.mean(arr1))
```

```python
print("Maximum Value:", np.max(arr1))
```

```python
print("Adding 10 to every element:")
print(arr1 + 10)
```
